```python
colors = cyc('rgb')
targets = range(len(lir().target_names))
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
for i, c, label in zip(targets, colors, lir().target_names):
    ax.scatter(newIris[lir().target == i, 0],
                  newIris[lir().target == i, 1], 
                  c = c, label = label)
plt.legend()

plt.show()
```
